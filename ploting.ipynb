```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]

t = np.linspace(0, 1.5*np.pi, 2500)
y = np.sin(t**2)+np.random.random(2500)*.6
df = pd.DataFrame(y)
# plt.plot(df[0], 'lightblue', df[0].rolling(10).mean(), 'b')

smoothing=2
dqn = pd.read_csv('csv_data/DQN.csv')
dueling_dqn = pd.read_csv('csv_data/dueling_DQN.csv')
plt.plot(dqn.Value, 'lavender')
plt.plot(dueling_dqn.Value, 'mistyrose')
plt.plot(dqn.Value.rolling(smoothing).mean(), 'royalblue', label='DQN')
plt.plot(dueling_dqn.Value.rolling(smoothing).mean(), 'orange', label='dueling-DQN')


plt.title("")
plt.xlabel("Episode")
plt.ylabel("Score")
plt.legend(loc='lower right')
```
