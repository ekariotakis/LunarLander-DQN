```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]


smoothing=15
dqn = pd.read_csv('csv_data/DQN.csv')
dueling_dqn = pd.read_csv('csv_data/dueling_DQN.csv')
plt.plot(dqn.Step, dqn.Value, 'lavender')
plt.plot(dueling_dqn.Step, dueling_dqn.Value, 'mistyrose')
plt.plot(dqn.Step, dqn.Value.rolling(smoothing).mean(), 'royalblue', label='DQN')
plt.plot(dueling_dqn.Step, dueling_dqn.Value.rolling(smoothing).mean(), 'orange', label='dueling-DQN')


plt.title("Comparison of Smoothed Score", fontsize=16)
plt.xlabel("Episode", fontsize=16)
plt.ylabel("Score", fontsize=16)
plt.legend(loc='lower right', fontsize=16)
plt.grid(color = 'lightgray', linestyle = '--', linewidth = 0.5)
plt.savefig('figures/smoothed_score.eps', format='eps')
```
